````python
from datasets import load_dataset

# 허깅페이스 데이터셋 로드 - 테스트용 (질의 데이터셋)
dataset = load_dataset('commonsense_qa')

# 데이터셋을 50%로 분할 - cuda 메모리 문제 이슈
split_dataset = dataset['train'].train_test_split(test_size=0.5)

# 분할된 데이터셋 할당
train_subset = split_dataset['train']
validation_subset = split_dataset['test']
````

````python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3-8B-Instruct")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Meta-Llama-3-8B-Instruct")
````

````python
# 패딩 토큰 추가
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': tokenizer.eos_token})

# 데이터 전처리 함수
def preprocess_function(examples):
    inputs = []
    labels = []
    for question, choices, answer in zip(examples['question'], examples['choices']['text'], examples['answerKey']):
        inputs.append([f"{question} {choice}" for choice in choices])
        labels.append(choices.index(answer))

    tokenized_inputs = tokenizer(
        inputs,
        truncation=True,
        padding='max_length',
        max_length=128,
        is_split_into_words=True
    )

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

# 데이터 전처리 적용
encoded_dataset = dataset.map(preprocess_function, batched=True, remove_columns=dataset['train'].column_names)

# 훈련 및 검증 데이터셋 설정
train_dataset = encoded_dataset['train']
eval_dataset = encoded_dataset['validation']
````

````python
# 로라 설정
from transformers import AutoTokenizer, TrainingArguments, Trainer
from peft import LoraConfig, PeftModel, get_peft_model

'''lora_config = LoraConfig(
    task_type="CAUSAL_LM",
    init_lora_weights="kaiming_uniform",
    target_modules="all-linear"
)'''

lora_config = LoraConfig(
    task_type="CAUSAL_LM", inference_mode=False, r=8, lora_alpha=32, lora_dropout=0.1
)

lora_model = PeftModel(model, lora_config)
````

````python
# 도라 설정
dora_config = LoraConfig(
    task_type="CAUSAL_LM",
    use_dora=True,
    inference_mode = False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.1 # 다 똑같이 설정
)

dora_model = PeftModel(model, dora_config)
````

````python
# 훈련 설정
training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/Colab Notebooks/results',
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    logging_dir='./logs',
    logging_steps=50,
    no_cuda=True,
    remove_unused_columns=False,
)
````

````python
from transformers import Trainer, TrainingArguments

# 훈련 함수 설정
def train_model(model, training_args, train_dataset, eval_dataset):
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        tokenizer=tokenizer,
    )
    trainer.train()
    return trainer

lora_trainer = train_model(lora_model, training_args, train_subset, validation_subset)
dora_trainer = train_model(dora_model, training_args, train_subset, validation_subset)
````

````python
# 성능 평가

def evaluate_model(trainer, dataset):
    metrics = trainer.evaluate(eval_dataset=dataset["validation"])
    return metrics

lora_metrics = evaluate_model(lora_trainer, dataset)
dora_metrics = evaluate_model(dora_trainer, dataset)

print("LoRA Metrics:", lora_metrics)
print("DoRA Metrics:", dora_metrics)
````
